## 参考答案与解析

````python
import math

def get_lr(current_step, total_steps, lr):
    return lr * (0.1 + 0.45 * (1 + math.cos(math.pi * current_step / total_steps)))
````

### 解析

1. 没有 warmup；这是项目当前实现，不是通用最佳实践。
2. total_steps 不能为 0。

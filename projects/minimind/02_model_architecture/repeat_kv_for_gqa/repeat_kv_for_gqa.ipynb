## 参考答案与解析

````python
import numpy as np

def repeat_kv(x, n_rep):
    x = np.asarray(x)
    if n_rep == 1:
        return x
    b, s, h, d = x.shape
    return np.repeat(x[:, :, :, None, :], n_rep, axis=3).reshape(b, s, h * n_rep, d)
````

### 解析

1. 这里重复的是 KV，不是 Q。
2. 真实 attention 前还会 transpose 到 `[B, H, S, D]`。

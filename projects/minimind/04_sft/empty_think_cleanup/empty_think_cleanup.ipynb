## 参考答案与解析

````python
def cleanup_empty_think(prompt, keep_empty=False):
    pattern = "<think>\n\n</think>\n\n"
    if keep_empty:
        return prompt
    return prompt.replace(pattern, "")
````

1. 真实代码里 keep/remove 是随机比例控制。
